### **Objective**

**Green Destinations is a well known travel agency. The HR Director has recently noticed an increase in employees leaving (attrition). She would like to figure out any trends or patterns. She has surveyed the staff of Green Destinations and provided you with the data. She would like to know what the attrition rate is (% of people who have left). She would also like to know if factors like age, years at the company and income play a part in
determining if people will leave or not.**

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv(r"C:\Users\DELL\Downloads\greendestination (1).csv")
#data.columns = data.columns.str.strip()  # Removes extra spaces

# Convert 'Attrition' to a binary column for analysis
data['Attrition_Binary'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculate the overall attrition rate
total_employees = data.shape[0]
attrition_count = data[data['Attrition'] == 'Yes'].shape[0]
attrition_rate = (attrition_count / total_employees) * 100

# Print the attrition rate
print(f"The attrition rate is approximately {attrition_rate:.2f}%")

# Plot the relationship between attrition and key factors
plt.figure(figsize=(20, 10))

# Histogram for Age vs Attrition
plt.subplot(2, 3, 1)
sns.histplot(data=data, x='Age', hue='Attrition', kde=True, multiple='stack', bins=20)
plt.title('Age vs Attrition')

# Histogram for YearsAtCompany vs Attrition
plt.subplot(2, 3, 2)
sns.histplot(data=data, x='YearsAtCompany', hue='Attrition', kde=True, multiple='stack', bins=20)
plt.title('Years at Company vs Attrition')

# Histogram for MonthlyIncome vs Attrition
plt.subplot(2, 3, 3)
sns.histplot(data=data, x='MonthlyIncome', hue='Attrition', kde=True, multiple='stack', bins=20)
plt.title('Monthly Income vs Attrition')

# Job Role vs. Attrition (using a count plot)
plt.subplot(2, 3, 4)
sns.countplot(x='JobRole', hue='Attrition', data=data)
plt.title('Job Role vs Attrition')
plt.xticks(rotation=45)

# Education Level vs. Attrition (using a count plot)
plt.subplot(2, 3, 5)
sns.countplot(x='Education', hue='Attrition', data=data)
plt.title('Education Level vs Attrition')

# Work-Life Balance vs. Attrition (using a count plot)
plt.subplot(2, 3, 6)
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=data)
plt.title('Work-Life Balance vs Attrition')

plt.tight_layout()
plt.show()

# Correlation analysis for numeric factors
correlation_matrix = data[['Age', 'YearsAtCompany', 'MonthlyIncome', 'Attrition_Binary']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Heatmap for correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Additional analysis for categorical factors
print("\nCategorical Analysis:\n")
categorical_factors = ['JobRole', 'Education', 'WorkLifeBalance', 'Department', 'MaritalStatus']
for factor in categorical_factors:
    attrition_counts = data.groupby(factor)['Attrition_Binary'].mean().sort_values(ascending=False)
    print(f"\nAverage Attrition Rate by {factor}:\n")
    print(attrition_counts)

# Conclusion and analysis:
print("\nAnalysis:\n")
print("1. The attrition rate at Green Destinations is approximately 16.12%, indicating a moderate level of turnover.")
print("2. Age appears to have some variation between employees who left and those who stayed. Further tests could determine if older or younger groups are more affected.")
print("3. Years at the company shows that employees with shorter or moderately longer tenures might be at risk, suggesting varied tenure impact.")
print("4. Monthly income may indicate certain salary levels are correlated with higher attrition, needing additional context to confirm.")
print("5. Job role, education level, and work-life balance show different average attrition rates. Certain job roles and lower work-life balance scores might be more prone to attrition.")
print("6. Departments and marital status could further indicate trends where specific groups have higher attrition rates.")


# Conclusion and further recommendations
print("\nConclusion and Recommendations:")
print("1. The attrition rate is approximately 16.12%, indicating a moderate level of turnover.")
print("2. Age and attrition show a weak negative correlation, suggesting older employees might be less likely to leave.")
print("3. Employees with shorter or moderately longer tenures may be at higher risk of leaving, suggesting tenure plays a role in attrition.")
print("4. Monthly income seems weakly correlated with attrition. Further analysis should explore salary bands to understand the impact.")
print("5. Job role, education level, and work-life balance have significant impacts on attrition rates, requiring tailored strategies for each.")
print("6. Departments and marital status further contribute to attrition trends, and focusing on these aspects could reduce turnover.")
```
